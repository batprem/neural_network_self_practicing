# Import modules

```python
import pandas as pd
from keras.layers import (Input, Concatenate, concatenate,
                          BatchNormalization, Conv2D, Flatten, Dense)

from keras.models import Model, load_model
import tensorflow as tf
```

# Load data

```python
dataset = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
```

# Preprocessing

```python
dataset = dataset.sample(dataset.shape[0], random_state=7)
```

```python
dataset
```

```python
Train, Test = dataset.iloc[0:100], dataset.iloc[100:]
```

```python
X_train, Y_train_precessed = Train.drop(['variety'], axis=1), Train[['variety']]
X_test, Y_test = Test.drop(['variety'], axis=1), Test[['variety']]
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_col = ['variety']
enc= OneHotEncoder(handle_unknown='ignore')
enc.fit(Y_train)
transform_data = enc.transform(Y_train)
```

```python
Y_train_precessed = pd.DataFrame(
                transform_data.toarray(),
                index=Y_train.index, # Set index
                columns=enc.get_feature_names(cat_col) # Set columns name
            )
transform_data = enc.transform(Y_test)
Y_test_processed = pd.DataFrame(
                transform_data.toarray(),
                index=Y_test.index, # Set index
                columns=enc.get_feature_names(cat_col) # Set columns name
            )
```

```python
pd.concat([Y_train, Y_train_precessed], axis=1)
# Then, we're going to use Y_train_precessed instead of Y_train
```

```python
tf.keras.backend.set_floatx('float64')
class IrisModel(Model):
    def __init__(self):
        super(IrisModel, self).__init__()
        self.dense_8 = Dense(8, activation='sigmoid')
        self.dense_8 = Dense(20, activation='sigmoid')
        self.dense_output = Dense(3, activation='sigmoid')
    def call(self, input_layer):
        hidden = self.dense_8(input_layer)
        hidden = self.dense_8(input_layer)
        output = self.dense_output(hidden)
        return output
    def summary(self):
        x = Input(shape=X_train.shape[1:])
        model = Model(inputs=[x], outputs=self.call(x))
        return model.summary()
    
model = IrisModel()

model.compile(loss=tf.keras.losses.BinaryCrossentropy(), optimizer='adam', metrics=['accuracy'])
hist = model.fit(X_train, Y_train_precessed , epochs=200)
```
